```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/Dataset/zomato.csv')
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
df.describe()
```

```python
online_order = df['online_order'].value_counts()
```

```python
# Plotting the graph
plt.figure(figsize=(8, 6))
online_order.plot(kind='bar', color=['skyblue', 'orange'], edgecolor='black')

# Adding title and labels
plt.title('Restaurants Accepting Online Orders', fontsize=14)
plt.xlabel('Online Order Acceptance', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)

# Display the counts on top of each bar
for i, count in enumerate(online_order):
    plt.text(i, count + 50, str(count), ha='center', fontsize=10, color='black')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
book_table = df['book_table'].value_counts()
```

```python
# Plotting the graph
plt.figure(figsize=(8, 6))
book_table.plot(kind='bar', color=['lightcoral', 'lightgreen'], edgecolor='black')

# Adding title and labels
plt.title('Restaurants Providing Table Booking', fontsize=14)
plt.xlabel('Table Booking Availability', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)

# Display the counts on top of each bar
for i, count in enumerate(book_table):
    plt.text(i, count + 50, str(count), ha='center', fontsize=10, color='black')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the top 25 restaurants by number of outlets
plt.figure(figsize=(12,8))
val = df['name'].value_counts()[:25]  # Considering for the top 25 restaurants
sns.barplot(x=val, y=val.index, palette="Set1")
plt.title("Famous Restaurants in the City of Mumbai")
plt.xlabel("Number of Outlets")
plt.show()  # Make sure the first plot is displayed

# Sorting the dataframe by 'rate' in descending order to get top-rated restaurants
top_restaurants = df[['name', 'rate']].sort_values(by='rate', ascending=False).head(10)

# Plotting the top 10 restaurants by rating
plt.figure(figsize=(10, 6))
plt.barh(top_restaurants['name'], top_restaurants['rate'], color='skyblue', edgecolor='black')

# Adding title and labels
plt.title('Top 10 Restaurants by Rating', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Restaurant Name', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Plotting the top 10 restaurants by number of outlets
plt.figure(figsize=(12,8))
val = df['name'].value_counts()[:10]  # Considering for the top 25 restaurants
sns.barplot(x=val, y=val.index, palette="Set1",legend=False)
plt.title("Famous Restaurants in the City of Bengaluru")
plt.xlabel("Number of Outlets")
plt.show()  # Make sure the first plot is displayed
```
