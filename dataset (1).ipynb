Data cleaning

```python
df = df.replace('?',np.NAN)
df
```

```python
df.isnull().sum()
```

Missing Data

```python
data = df['normalized-losses'].loc[~df['normalized-losses'].isna()]

mean = data.astype(int).mean()

df['normalized-losses'] = df['normalized-losses'].replace(np.nan,mean).astype(int)
df['normalized-losses'].head()
```

```python
df['normalized-losses'].dtypes
```

```python
df.head()
```

```python
df.columns
```

```python
df['normalized-losses'].unique()
```

```python
df['price'].unique() # ? missing value
```

```python
data = df['price'].loc[~df['price'].isna()]
mean = data.astype(int).mean()
df['price'] = df['price'].replace(np.nan,mean).astype(int)
df['price'].head()
```

```python
df['price'].unique()
```

```python
data = df['horsepower'].loc[~df['horsepower'].isna()]
mean = data.astype(int).mean()
df['horsepower'] = df['horsepower'].replace(np.nan,mean).astype(int)
df['horsepower'].head()
```

```python
data = df['peak-rpm'].loc[~df['peak-rpm'].isna()]
mean = data.astype(int).mean()
df['peak-rpm'] = df['peak-rpm'].replace(np.nan,mean).astype(int)
df['peak-rpm'].head()
```

```python
df.head(20)
```

```python
# Replace the non-numeric value to null and convert the datatype

df['bore'] = pd.to_numeric(df['bore'],errors='coerce')
df.dtypes
```

```python
# Replace the non-number value to null and convert the datatype
df['stroke'] = pd.to_numeric(df['stroke'],errors='coerce')
df.dtypes
```
